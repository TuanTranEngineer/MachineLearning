```python
import sys
sys.path.insert(0, './../Libs/')
import numpy as np
from pca import PCA
import matplotlib.pyplot as plt
```

```python
cov = np.array([[1, 0.9],
                [0.9, 1]])
mean = np.array([1, 1])
```

```python
x0, x1 = np.random.multivariate_normal(mean, cov, 100).T

X = np.hstack((x0.reshape(-1, 1), x1.reshape(-1, 1)))
pca = PCA(X).fit()
plt.figure(figsize=(5, 5))
plt.scatter(x0, x1)
plt.scatter(pca.Xmean[0], pca.Xmean[1], c = 'r')

vector0 = pca.eigenVectors[0]
vector1 = pca.eigenVectors[1]
c0 = pca.Xmean.dot(vector0).item()
c1 = pca.Xmean.dot(vector1).item()
xline = np.linspace(-2, 4, 10)
a0 = vector0[0]
a1 = vector1[0]
b0 = vector0[1]
b1 = vector1[1]

y0 = (c0 - a0*xline)/b0
y1 = (c1 - a1*xline)/b1

plt.plot(xline, y0)
plt.plot(xline, y1)
plt
plt.xlim(-2, 4)
plt.ylim(-2, 4)
```
