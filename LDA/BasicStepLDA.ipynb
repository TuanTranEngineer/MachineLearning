```python
import numpy as np
import sys
sys.path.insert(0, './../Libs/')
from lda import TwoClassesLDA, LDA
import sklearn.datasets as skdata
import matplotlib.pyplot as plt
```

```python
NsamplesEachClass = 50
sampleNoise = 0.3
seed = 0
centers = [
    [0, 0],
    [1, 2]
]
```

```python
X, t = skdata.make_blobs(n_samples = [NsamplesEachClass, NsamplesEachClass], 
                         cluster_std = sampleNoise, random_state = seed, shuffle = False, centers = centers)

plt.figure(figsize = (10, 10))
plt.scatter(X[:NsamplesEachClass, 0], X[:NsamplesEachClass, 1], c = 'r', label = "class -1")
plt.scatter(X[NsamplesEachClass:, 0], X[NsamplesEachClass:, 1], c = 'g', label = "class 1")
plt.xlim((-1, 2))
# plt.ylim((-1, 2))
plt.legend()
```

```python
lda = TwoClassesLDA().fit(X, t)
w = lda.getw()
print(w)
slope = w[0]/w[1]
xplot = np.linspace(-1, 2, 100)
yplot = xplot*slope

plt.figure(figsize = (10, 10))
plt.scatter(X[:NsamplesEachClass, 0], X[:NsamplesEachClass, 1], c = 'r', label = "class -1")
plt.scatter(X[NsamplesEachClass:, 0], X[NsamplesEachClass:, 1], c = 'g', label = "class 1")
plt.plot(xplot, yplot)
plt.xlim((-1, 2))
# plt.ylim((-1, 2))
plt.legend()
```

```python
lda = LDA().fit(X, t)
w = lda.getw()
print(w)
w = w[:, 1]
slope = w[0]/w[1]
xplot = np.linspace(-1, 2, 100)
yplot = xplot*slope

plt.figure(figsize = (10, 10))
plt.scatter(X[:NsamplesEachClass, 0], X[:NsamplesEachClass, 1], c = 'r', label = "class -1")
plt.scatter(X[NsamplesEachClass:, 0], X[NsamplesEachClass:, 1], c = 'g', label = "class 1")
plt.plot(xplot, yplot)
plt.xlim((-1, 2))
plt.ylim((-1, 2))
plt.legend()
```
